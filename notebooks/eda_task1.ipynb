```python

# STEP 2: Basic info
print("‚úÖ Basic Info:")
display(df.info())

# STEP 3: Check missing values
print("\nüîç Missing Values:")
missing = df.isnull().sum()
display(missing[missing > 0])

# STEP 4: Convert TransactionMonth to datetime
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'], errors='coerce')

# STEP 5: Drop duplicates
df = df.drop_duplicates()

# STEP 6: Strip whitespace from object (string) columns
object_cols = df.select_dtypes(include='object').columns
df[object_cols] = df[object_cols].apply(lambda x: x.str.strip())

# STEP 7: Fill missing numerical values with median
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# STEP 8: Fill missing categorical values with mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

# STEP 9: Re-check missing values
print("\n‚úÖ Remaining Missing Values After Cleaning:")
display(df.isnull().sum()[df.isnull().sum() > 0])

# STEP 10: Preview cleaned data
print("\n‚úÖ Sample Cleaned Data:")
display(df.head())

# STEP 11: Save cleaned data (optional)
output_path = "cleaned_data/insurance_claims_cleaned.csv"
import os
os.makedirs("cleaned_data", exist_ok=True)
df.to_csv(output_path, index=False)
print(f"\n‚úÖ Cleaned data saved to: {output_path}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plot style
sns.set(style='whitegrid')
plt.rcParams.update({'axes.titlesize': 14, 'axes.labelsize': 12})

# Create reports directory if missing
os.makedirs('../reports', exist_ok=True)

# Numerical Variables
fig, axs = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['TotalPremium'], bins=30, kde=True, ax=axs[0], color='skyblue')
axs[0].set_title('Distribution of Total Premium')
axs[0].set_xlabel('Total Premium')

sns.histplot(df['TotalClaims'], bins=30, kde=True, ax=axs[1], color='salmon')
axs[1].set_title('Distribution of Total Claims')
axs[1].set_xlabel('Total Claims')

plt.tight_layout()
plt.savefig('../reports/premium_claims_distribution.png')
plt.show()

# Categorical Variables
fig, axs = plt.subplots(1, 3, figsize=(16, 5))
sns.countplot(data=df, x='Province', ax=axs[0], palette='Blues')
axs[0].set_title('Province Distribution')
axs[0].tick_params(axis='x', rotation=45)

sns.countplot(data=df, x='Gender', ax=axs[1], palette='coolwarm')
axs[1].set_title('Gender Distribution')

sns.countplot(data=df, x='VehicleType', ax=axs[2], palette='Set3')
axs[2].set_title('Vehicle Type Distribution')
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('../reports/categorical_distributions.png')
plt.show()
```
